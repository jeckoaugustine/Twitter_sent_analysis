```python
twit_train_data['len']=twit_train_data['tweet'].apply(lambda x: len(x))
```

```python
twit_train_data['clean_d']= twit_train_data['tweet'].apply(lambda x:' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",x).split()))
```

```python
twit_train_data['len']=twit_train_data['clean_d'].apply(lambda x: len(x))
twit_train_data['len'].describe()
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
twit_train_data['len'].hist(bins=20,ax=ax)
fig.tight_layout()
```

```python
word_freq=pd.DataFrame(data=twit_train_data['clean_d'].str.split(expand=True).stack().value_counts())
```

```python
if len(word_freq.columns) ==1:
    word_freq.reset_index(inplace=True)
word_freq.head(10)
```

```python
word_freq.columns=['words','count']
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
word_freq['count'].hist(bins=5,ax=ax)
fig.tight_layout()
```
